####***12.3. Evaluación de Outliers:***

Identificamos y tratamos outliers si es necesario. En analisis anterior, se había catalogado a Qatar como outlier, detectado en el punto 4.5. Vamos a ver si es correcto, para ello aplico el método de IQR

```python
# Análisis de outliers en "salary_usd"
col = "salary_in_usd"  # Columna a analizar

print(f"\nAnálisis de outliers para columna: {col}")

# Crear boxplot para "salary_usd"
sns.boxplot(y=col, data=df_jobs)
plt.xticks(rotation=90)  # Rotar etiquetas del eje X en 90 grados
plt.show()

# Identificar outliers usando IQR
q1 = df_jobs[col].quantile(0.25)
q3 = df_jobs[col].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Buscar valores que superan el límite superior
outlier_values = df_jobs[df_jobs[col] > upper_bound][col].values
if len(outlier_values) > 0:
    print(f"Valores outliers encontrados: {outlier_values}")
else:
    print("No se encontraron valores outliers.")
```

El valor de Qatar era 300000, y como vemos en esta etapa los outliers son superiores, por lo tanto descarto la hipótesis planteada en el punto 4.5

Lo muestro tambien con el método z-score, definiendo valor >3
Muestro el valor de los outliers para verificar que el valor de Qatar no lo era.
Además, agrego información adicional, como la cantidad de outliers detectados y que porcentaje de los datos representa

```python
# Análisis de outliers en "salary_in_usd"
col = "salary_in_usd"  # Columna a analizar

print(f"\nAnálisis de outliers para columna: {col}")

# Calcular z-scores
zscores = np.abs(df_jobs[col] - df_jobs[col].mean()) / df_jobs[col].std()

# Identificar outliers usando un umbral (por ejemplo, 3)
outlier_values = df_jobs.loc[zscores > 3, col].values
if len(outlier_values) > 0:
    print(f"Valores outliers encontrados (z-score > 3): {outlier_values}")
else:
    print("No se encontraron valores outliers con z-score > 3.")

# Contar la cantidad de outliers
cantidad_outliers = len(outlier_values)
print(f"Cantidad de outliers encontrados: {cantidad_outliers}")

# Calcular el porcentaje de outliers
porcentaje_outliers = (cantidad_outliers / len(df_jobs)) * 100
print(f"Porcentaje que representan los outliers: {porcentaje_outliers:.2f}%")

# Visualización (modificado para usar 'loc' en la selección)
plt.figure(figsize=(12, 6))
sns.boxplot(y=col, data=df_jobs, showmeans=True)

# Marcar outliers con z-score > 3 en rojo
for i, row in df_jobs.iterrows():
    if zscores.loc[i] > 3:
        plt.scatter(i, row[col], color="red")

plt.title(f"Distribución de {col} con outliers marcados (z-score > 3)")
plt.show()

```
